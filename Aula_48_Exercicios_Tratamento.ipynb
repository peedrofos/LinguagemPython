```python
import pandas as pd

# Iniciar importando os dados da planilha:

df = pd.read_excel('vendas.xlsx')
#print(df.isnull)
#print(df)

# 1.Padronizar a coluna &#39;categoria&#39; para minúsculas.
df['categoria'] = df['categoria'].str.lower() 
#print(df)

# 2.Remover registros onde a quantidade está ausente
df = df[df['quantidade'].notna()]
#print(df)

# 3. Remover espaços extras e padronizar com primeira letra maiúscula (estilo título)
df['categoria'] = df['categoria'].str.strip().str.title()
df['produto'] = df['produto'].str.strip().str.title()
df['loja'] = df['loja'].str.strip().str.title()
df.columns = df.columns.str.strip().str.title()
#print(df)

# 4. Garantir que a coluna data seja datetime
df['Data'] = pd.to_datetime(df['Data'])

# 5. Calcular valor unitário real das vendas válidas
df['Valor_Unitario'] = df['Valor_Venda'] / df['Quantidade']
#print(df)

# 6. Calcular valor médio unitário por produto (sem considerar nulos)
media_por_produto = df.groupby('Produto', as_index=False)['Valor_Unitario'].mean().round(2)
#print(media_por_produto)

# 7. Agrupar e calcular média dos valores unitários
media_por_produto = df.groupby('Produto', as_index=False)['Valor_Unitario'].mean()
media_por_produto['Valor_Unitario'] = media_por_produto['Valor_Unitario'].round(2) 
#print(media_por_produto)

# 8. Preencher valor_unitario ausente
media_dic = dict(zip(media_por_produto['Produto'], media_por_produto['Valor_Unitario']))
df['Valor_Unitario'] = df['Valor_Unitario'].fillna(df['Produto'].map(media_dic))
#print(media_dic)

# 9. Recalcular valor_venda
df['Valor_Venda'] = df['Valor_Unitario'] * df['Quantidade']
#print(df)

# 10. Salvar o novo DataFrame limpo em um novo arquivo Excel
df.to_excel("vendas_tratadas.xlsx", index=False)
print("Tratamento finalizado e salvo como 'vendas_tratadas.xlsx'")
```
